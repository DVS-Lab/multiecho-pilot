```python
# Print all subs from sourcedata dir with more than one localizer
base_path = "/ZPOOL/data/sourcedata/sourcedata/rf1-sequence-pilot/"

# List all subject directories
subject_dirs = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]

for subject_dir in subject_dirs:
    subject_path = os.path.join(base_path, subject_dir, subject_dir, "scans")
    #print("Subject path:", subject_path)
    
    if not os.path.exists(subject_path):
        #print(f"Path does not exist: {subject_path}")
        continue

    result = subprocess.run(['ls', '-lrt'], cwd=subject_path, capture_output=True, text=True)
    lines = result.stdout.split('\n')
    
    localizer_count = 0
    localizer_scans = []

    for line in lines:
        if line:
            last_column = line.split()[-1]
            if "localizer" in last_column:
                localizer_count += 1
                localizer_scans.append(last_column)
    
    if localizer_count > 1:
        print("Subject path:", subject_path)
        for scan in localizer_scans:
            print(scan)
        print("----------------------------------------------------\n")
```

```python
# Main script kernel

# Read subject IDs from a text file
subjects_file = '/ZPOOL/data/projects/multiecho-pilot/code/sublist-deriv.txt'
with open(subjects_file, 'r') as f:
    subjects = f.read().splitlines()

# Process each subject
for subject in subjects:
    print(f"Processing subject {subject}...")
    
    # Get the subject directories
    subject_dirs = glob.glob(f"/ZPOOL/data/sourcedata/sourcedata/rf1-sequence-pilot/Smith-SRA-{subject}/Smith-SRA-{subject}/scans")
    
    # Initialize the scan list
    scan_list = []
    
    # Populate the scan list with ls -lrt command
    for subject_dir in subject_dirs:
        result = subprocess.run(['ls', '-lrt', subject_dir], capture_output=True, text=True)
        scan_list = result.stdout.splitlines()[1:] # Ignore the first line (total)
    #print("/n".join(scan_list)) # Print the scan list for debugging purposes
    
    # Initialize vars
    fmap_count = 0
    fmap_sequence = 0
    prev_fmap_run = None
    functional_runs = []
    fmap_runs = {}
    
    # Process each line in the scan list
    for line in scan_list:
        last_column = line.split()[-1]
        
        # Check if the line contains a fieldmap scan
        if "fmap" in last_column:
            fmap_sequence += 1
            if fmap_sequence == 1:
                fmap_count += 1
                fmap_runs[f"fmap_run-{fmap_count}"] = []
            prev_fmap_run = f"fmap_run-{fmap_count}"
            #print(f"{last_column} is {prev_fmap_run}")
            
            # Reset fmap_sequence after the third fmap
            if fmap_sequence == 3:
                fmap_sequence = 0
        
        # Check if the line contains a functional scan
        elif "CMRR" in last_column:
            functional_runs.append(last_column)
            if prev_fmap_run:
                mb_index = last_column.find("MB")
                me_index = last_column.find("ME")
                mb_value = last_column[mb_index:mb_index + 3].lower() # Extract MB value and convert to lowercase
                me_value = last_column[me_index:me_index + 3].lower() # Extract ME value and convert to lowercase
                mb_me_combination = f"{mb_value}{me_value}"
                if mb_me_combination not in fmap_runs[prev_fmap_run]:
                    fmap_runs[prev_fmap_run].append(mb_me_combination)
                #print(f"Intended for {prev_fmap_run}: {last_column}")
                #print(f"MB={mb_value[2]}, ME={me_value[2]}")
                #print(f"Current combinations for {prev_fmap_run}: {fmap_runs[prev_fmap_run]}")
                
        
        # Reset the sequence if the line is neither a fmap nor functional run:
        else:
            fmap_sequence = 0
    
    #print(f"Final fmap runs for subject {subject}: {fmap_runs}")
    #print("\n")

    # Create separate lists for each fmap_run
    fmap_run_lists = {}
    for fmap_run, mb_me_combinations in fmap_runs.items():
        fmap_run_lists[f"{fmap_run}_func"] = mb_me_combinations

    # Print the parsed lists for each fmap_run
    for fmap_run, combinations in fmap_run_lists.items():
        print(f"{fmap_run}: {combinations}")
        
    # Define the base directory for the JSON files
    json_base_dir = f"/ZPOOL/data/projects/multiecho-pilot/bids/sub-{subject}/fmap"
    
    # Update JSON files based on fmap_run_lists
    for fmap_run, mb_me_combinations in fmap_run_lists.items():
        # Skip if combinations is empty
        if not mb_me_combinations:
            continue
        
        # Extract run number
        run_number = fmap_run.split('-')[1].split('_')[0]

        # Define JSON file paths for the current fmap_run
        magnitude1_json = os.path.join(json_base_dir, f"sub-{subject}_acq-bold_run-{run_number}_magnitude1.json")
        magnitude2_json = os.path.join(json_base_dir, f"sub-{subject}_acq-bold_run-{run_number}_magnitude2.json")
        phasediff_json = os.path.join(json_base_dir, f"sub-{subject}_acq-bold_run-{run_number}_phasediff.json")

        # List of JSON files to update
        json_files = [magnitude1_json, magnitude2_json, phasediff_json]

        # Collect all intended functional runs
        intended_for = []

        for mb_me_combination in mb_me_combinations:
            mb = mb_me_combination[:3]  # Extract mb part
            me = mb_me_combination[3:]  # Extract me part
            
            if me == "me1":
                intended_for.append(f"func/sub-{subject}_task-sharedreward_acq-{mb}{me}_bold.nii.gz")
            elif me == "me4":
                for echo in range(1, 5):
                    intended_for.append(f"func/sub-{subject}_task-sharedreward_acq-{mb}{me}_echo-{echo}_part-phase_bold.nii.gz")
                    intended_for.append(f"func/sub-{subject}_task-sharedreward_acq-{mb}{me}_echo-{echo}_part-mag_bold.nii.gz")

        # Update each JSON file with the intended functional runs
        for json_file in json_files:
            #print(f"{json_file}")
            if os.path.exists(json_file):
                try:
                    with open(json_file, 'r') as f:
                        json_data = json.load(f)
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON file {json_file}: {e}")
                    with open(json_file, 'r') as f:
                        print(f.read()) # Print the content of the problematic JSON file
                    continue
                
                intended_for = []
                for comb in mb_me_combinations:
                    mb = comb.split('me')[0][2:]
                    me = comb.split('me')[1]
                    
                    if me == '1':
                        intended_for.append(f"func/sub-{subject}_task-sharedreward_acq-mb{mb}me1_bold.nii.gz")
                    elif me == '4':
                        intended_for.extend([
                            f"func/sub-{subject}_task-sharedreward_acq-mb{mb}me4_echo-1_part-mag_bold.nii.gz",
                            f"func/sub-{subject}_task-sharedreward_acq-mb{mb}me4_echo-1_part-phase_bold.nii.gz",
                            f"func/sub-{subject}_task-sharedreward_acq-mb{mb}me4_echo-2_part-mag_bold.nii.gz",
                            f"func/sub-{subject}_task-sharedreward_acq-mb{mb}me4_echo-2_part-phase_bold.nii.gz",
                            f"func/sub-{subject}_task-sharedreward_acq-mb{mb}me4_echo-3_part-mag_bold.nii.gz",
                            f"func/sub-{subject}_task-sharedreward_acq-mb{mb}me4_echo-3_part-phase_bold.nii.gz",
                            f"func/sub-{subject}_task-sharedreward_acq-mb{mb}me4_echo-4_part-mag_bold.nii.gz",
                            f"func/sub-{subject}_task-sharedreward_acq-mb{mb}me4_echo-4_part-phase_bold.nii.gz"
                            ])
                
                # Overwrite the 'IntendedFor' field
                #json_data['IntendedFor'] = intended_for

                # Print changes
                print(f"Updating {json_file}: IntendedFor -> {intended_for}")
                
                # Write back to the JSON file
                #with open(json_file, 'w') as f:
                #    json.dump(json_data, f, indent=4)
            else:
                print(f"JSON file {json_file} not found.")
    
print("Processing completed.")
```
