```python
for subject in subjects:
    subject_path = "/ZPOOL/data/sourcedata/sourcedata/rf1-sequence-pilot/Smith-SRA-" + subject + "/Smith-SRA-" + subject + "/scans"
    print("Subject path:", subject_path)
    result = subprocess.run(['ls', '-lrt'], cwd=subject_path, capture_output=True, text=True)
    lines = result.stdout.split('\n')
    
    fmap_count = 0
    fmap_sequence = 0
    prev_fmap_run = None
    functional_runs = []
    for line in lines:
        if line:
            last_column = line.split()[-1]
            if "fmap" in last_column:
                fmap_sequence += 1
                if fmap_sequence == 1:
                    fmap_count += 1
                prev_fmap_run = f"fmap_run-{fmap_count}"
                print(f"{last_column} is {prev_fmap_run}")
            elif "CMRR" in last_column:
                functional_runs.append(last_column)
                if prev_fmap_run:
                    mb_value = [int(s[2:]) for s in last_column.split("_") if "MB" in s][0]
                    me_value = [int(s[2:]) for s in last_column.split("_") if "ME" in s][0]
                    print(f"Intended for {prev_fmap_run}: {last_column} (MB={mb_value}, ME={me_value})")
            else:
                fmap_sequence = 0
    print("----------------------------------------------------\n")
```
